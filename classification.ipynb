```python
def build_tree(df):
    tree = {}
    for _, row in df.iterrows():
        category_id = row['category_id']
        parent_id = row['parent_id']
        if parent_id not in tree:
            tree[parent_id] = []
        tree[parent_id].append(category_id)
    return tree
```

```python
def print_tree(tree, current_node, indent=""):
    children = tree.get(current_node, [])
    for child in children:
        print(indent + str(child))
        print_tree(tree, child, indent + "  ")
```

```python
tree = build_tree(category_parent)
print_tree(tree, -1)
```
